# Introduction

This is an exploratory data analysis for Iris dataset taken from Kaggle which is best known for pattern recognition literature.
This is a multivariate dataset which explains about three related Iris flowers species. It contains 150 rows and 6 columns.
The three species in this data are Iris-Setosa, Iris-Versicolor and Iris-Virginica. It contains 50 sample of each species.
There are 4 features for each species to describe. They are:
1. Petal Length in cm
2. Petal Width in cm
3. Sepal Length in cm
4. Sepal Width in cm

The aim of this project is to predict the type of species based on its features.

# Data Visualization

```python
iris["SepalLengthCm"].plot(kind="hist", bins=50)
plt.xlabel("Sepal Length in Cms")
```

```python
iris["SepalLengthCm"].plot(kind="line")
plt.ylabel("Sepal Length in Cms")
```

```python
sns.set_style('darkgrid')
sns.FacetGrid(iris, hue = 'Species', size = 10)\
.map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm')\
.add_legend()
plt.show()
```

Observations:
1. Using Petal_Length and Petal_Width features, we can identify Setosa, Versicolor and Virginica fairly
2. Overlaping exists between Versicolor & Virginica.
3. Graph shows that Petal features are best contributor for Iris Species.

```python
sns.set_style('darkgrid')
sns.FacetGrid(iris, hue = 'Species', size = 10)\
.map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm')\
.add_legend()
plt.show()
```

Observations:
1. Using Sepal_Length and Sepal_Width features, we can only identify Setosa flower from other flowers
2. Differentiating Versicolor and Virginica is difficult as they have considerable overlap
3. Hence, Sepal_Length and Sepal_Width features work well for Setosa

```python
plt.close()
sns.pairplot(iris, hue = 'Species', size = 2, diag_kind = 'hist')
plt.show()
```
